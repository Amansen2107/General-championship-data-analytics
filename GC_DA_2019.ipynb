```python
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder as le
a= train['Congestion_Type']
le = le()
```

```python
label = le.fit_transform(a)
```

```python
df = pd.DataFrame(label)
```

```python
train = train.drop(['Congestion_Type'],axis=1)
```

```python
train = pd.concat([train,df],axis=1)
```

```python
train['Congestion_Type']=train[0] 
train = train.drop([0],axis=1)
```
